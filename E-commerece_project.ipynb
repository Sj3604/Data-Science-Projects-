```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = "plotly_white"

# encoding is put here because with the help of this it can read the special characters which are there in the file
# otherwise you won't be able to read the file properly
data = pd.read_csv("/content/Sample - Superstore (1).csv",encoding = "latin-1")
data
```

```python
# inorder to get the basic statistics we will use the describe function
data.describe()
```

```python
# using the info function inorder to get the more information about the dataset
data.info()
```

```python
# converting the date columns to a more suitable format from object to date-time
data["Order Date"] = pd.to_datetime(data["Order Date"])
data["Ship Date"] = pd.to_datetime(data["Ship Date"])
data.info()
```

```python
# inorder to check the working of the new data type
data.head()
```

```python
# we are doing this so that we can track the sales for the month and the year as well
data['Order Month'] = data['Order Date'].dt.month
data['Order Year'] = data['Order Date'].dt.year
data['Order Day Of Week'] = data['Order Date'].dt.dayofweek
data.head()
```

```python
# Monthly Sales Analysis

sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
fig = px.line(sales_by_month, x='Order Month', y='Sales', title='Monthly Sales Analysis')
fig.show()



#conclusion 1 : the highest sales recorded in November and lowest in February
```

```python
# Sales by Category

sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
fig = px.bar(sales_by_category, x='Category', y='Sales', title='Sales by Category')
fig.show()

# conclusion 2 :the category which was sold the most was technology and least was office supplies
```

```python
# Sales Analysis by the Sub-Categories

sales_by_sub_category = data.groupby('Sub-Category')['Sales'].sum().reset_index()
fig = px.bar(sales_by_sub_category, x='Sub-Category', y='Sales', title='Sales Analysis by Sub-Category')
fig.show()


# conclusion 3:the phones are most sold sub category and fasterners are the least sold sub category
```

```python
# Monthly Profit Analysis
data.head()

monthly_profit_analysis = data.groupby('Order Month')['Profit'].sum().reset_index()
fig = px.line(monthly_profit_analysis, x='Order Month', y='Profit', title='Monthly Profit Analysis')
fig.show()

# conclusion 4: the most profit was in the 12th month and the least was in 1st month
```

```python
# Profit by Category

profit_by_category = data.groupby('Category')['Profit'].sum().reset_index()
fig = px.bar(profit_by_category,x ='Category',y='Profit',title = 'Profit by Category')
fig.show()


# conclusion 5:the order of the profit by the categories is technology > office supplies > furniture
```

```python
# Profit by Sub-Category
profit_by_sub_category = data.groupby('Sub-Category')['Profit'].sum().reset_index()
fig = px.bar(profit_by_sub_category,x ='Sub-Category',y='Profit',title = 'Profit by Sub-Category')
fig.show()

# conclusion 6:the  most profit in sub category was of the copiers and the most loss was seen on the tables
#conclusion 7:even the phones was having the highest sales then also it did not earnt the msot profit
```

```python
# Sales and Profit-Customer Segment
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

color_palette = colors.qualitative.Pastel

fig = go.Figure()
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Sales'],
                     name='Sales',
                     marker_color=color_palette[0]))

fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Profit'],
                     name='Profit',
                     marker_color=color_palette[1]))

fig.update_layout(title='Sales and Profit Analysis by Customer Segment',
                  xaxis_title='Customer Segment', yaxis_title='Amount')

fig.show()



# conclusion 8:the most shoppin is done by the consumers and according to that they got the maximum profit by comparison
```

```python
# Sales to Profit Ratio
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
sales_profit_by_segment['Sales_to_Profit_Ratio'] = sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment', 'Sales_to_Profit_Ratio']])


# conclusion 9:the most profit ratio is for the consumers and least for the home office
```
